```python
import pandas as pd
import csv
import unicodedata
import spacy
import torch
```

```python
import wikipediaapi

# Definir un user_agent personalizado
user_agent = 'Usoacadémicoparatransformadores (monicagonzalezantolin1973@gmail.com)'

wiki_wiki = wikipediaapi.Wikipedia(
    language='en',
    extract_format=wikipediaapi.ExtractFormat.WIKI,
    user_agent=user_agent
)

# Función para obtener los nombres de una categoría específica
def print_categorymembers(categorymembers, level=0, max_level=1):
        for c in categorymembers.values():
            print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
            if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
                print_categorymembers(c.categorymembers, level=level + 1, max_level=max_level)

# Obtenemos la categoría deseada (en este caso, "20th-century women scientists")
cat = wiki_wiki.page("Category:20th-century women scientists")

# Miembros de la categoría
print_categorymembers(cat.categorymembers)
```

```python
cat = wiki_wiki.page("Category:20th-century women scientists")
cientificas_names = [c.title for c in cat.categorymembers.values()]

# Tenemos la lista, escribimos en el archivo CSV
with open('cientificas_names.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    for name in cientificas_names:
        writer.writerow([name])
```

```python
# Función para obtener detalles de una página de Wikipedia
def get_page_details(page_title):
    page = wiki_wiki.page(page_title)
    return {
        "title": page.title,
        "summary": page.summary,
        "url": page.fullurl
    }

# Cargar los nombres desde el archivo CSV
cientificas_details = []
with open('cientificas_names.csv', mode='r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        name = row[0]  # Asumiendo que cada fila contiene un nombre
        details = get_page_details(name)
        cientificas_details.append(details)

# Escribir los detalles recolectados en un nuevo archivo CSV
with open('cientificas_details.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Escribir los encabezados de la columna
    writer.writerow(["Title", "Summary", "URL"])

    # Escribir los detalles de cada científica en el CSV
    for details in cientificas_details:
        writer.writerow([details["title"], details["summary"], details["url"]])
```

```python
# Modificar la función para manejar valores que no son strings
def remove_non_ascii(text):
    if isinstance(text, str):  # Verifica si 'text' es una instancia de str
        return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    else:
        return text  # Si no es un string, simplemente devuelve el valor original

# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv('cientificas_details.csv')

# Aplicar la función de limpieza a la columna de resúmenes
df['Summary'] = df['Summary'].apply(remove_non_ascii)
df['Title'] = df['Title'].apply(remove_non_ascii)

# Guardar el DataFrame limpio de nuevo en un archivo CSV
df.to_csv('cientificas_details_clean.csv', index=False, encoding='utf-8-sig')
```

```python
# Cargar el modelo preentrenado de spaCy
nlp = spacy.load('en_core_web_sm')

# Función para extraer países o nacionalidades usando NER
def extract_country_or_nationality(summary):
    # Si el resumen es NaN (un número flotante en pandas), lo convertimos en una cadena vacía
    if isinstance(summary, float):
        return "Desconocido"
    doc = nlp(summary)
    for ent in doc.ents:
        if ent.label_ == "GPE" or ent.label_ == "NORP":
            return ent.text
    return "Desconocido"


df = pd.read_csv('cientificas_details_clean.csv')

# Se añade una nueva columna al DataFrame para los países
df['Country'] = df['Summary'].apply(extract_country_or_nationality)

# Verificar los resultados
print(df[['Title', 'Country']])

```

```python
# Funciones de búsqueda

def buscar_pais(nombre_cientifica):
    resultado = df[df['Title'] == nombre_cientifica]['Country']
    return resultado.iloc[0] if not resultado.empty else "Información no disponible"

def obtener_contribuciones(nombre_cientifica):
    resultado = df[df['Title'] == nombre_cientifica]['Summary']
    return resultado.iloc[0] if not resultado.empty else "Información no disponible"

def obtener_url(nombre_cientifica):
    resultado = df[df['Title'] == nombre_cientifica]['URL']
    return resultado.iloc[0] if not resultado.empty else "Información no disponible"

# Ejemplos de cómo usar las funciones
nombre_ejemplo = "Charlotte Anderson"
print(f"País de {nombre_ejemplo}: {buscar_pais(nombre_ejemplo)}")
print(f"Contribuciones de {nombre_ejemplo}: {obtener_contribuciones(nombre_ejemplo)}")
print(f"URL para más información sobre {nombre_ejemplo}: {obtener_url(nombre_ejemplo)}")
```

```python
def buscar_cientificas_por_pais_o_nacionalidad(termino):
    resultado = df[df['Country'].str.contains(termino, case=False, na=False)]['Title']
    if resultado.empty:
        return f"No se encontraron científicas importantes en {termino} o de nacionalidad {termino}"
    else:
        return resultado.tolist()

# Ejemplo de cómo usar la función
termino_busqueda = "Polish"  # Término que puede referirse al país Polonia o a la nacionalidad polaca
cientificas_en_termino = buscar_cientificas_por_pais_o_nacionalidad(termino_busqueda)
print(f"Científicas importantes en/de {termino_busqueda}: {cientificas_en_termino}")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Crear widgets para la entrada de datos
nombre_widget = widgets.Text(
    value='',
    description='Nombre:',
    disabled=False
)

pais_o_nacionalidad_widget = widgets.Text(
    value='',
    description='País/Nacionalidad:',
    disabled=False
)

# Botones para realizar la búsqueda
boton_buscar_nombre = widgets.Button(description="Buscar por Nombre")
boton_buscar_resumen = widgets.Button(description="Buscar Resumen")
boton_buscar_pais = widgets.Button(description="Buscar por País/Nacionalidad")

# Área de texto para mostrar resultados
area_resultados = widgets.Output()

# Eventos de los botones
def on_boton_buscar_nombre_clicked(b):
    with area_resultados:
        clear_output()
        nombre = nombre_widget.value
        pais = buscar_pais(nombre)
        contribuciones = obtener_contribuciones(nombre)
        url = obtener_url(nombre)

        print(f"País de {nombre}: {pais}")
        print(f"Contribuciones de {nombre}: {contribuciones}")
        if url != "Información no disponible":
            display(HTML(f"<a href='{url}' target='_blank'>URL para más información sobre {nombre}</a>"))
        else:
            print("URL para más información: Información no disponible")

def on_boton_buscar_resumen_clicked(b):
    with area_resultados:
        clear_output()
        nombre = nombre_widget.value
        resumen = obtener_contribuciones(nombre)
        print(f"Resumen de {nombre}: {resumen}")

def on_boton_buscar_pais_clicked(b):
    with area_resultados:
        clear_output()
        termino = pais_o_nacionalidad_widget.value
        cientificas = buscar_cientificas_por_pais_o_nacionalidad(termino)
        print(f"Científicas importantes en/de {termino}: {cientificas}")

boton_buscar_nombre.on_click(on_boton_buscar_nombre_clicked)
boton_buscar_resumen.on_click(on_boton_buscar_resumen_clicked)
boton_buscar_pais.on_click(on_boton_buscar_pais_clicked)

display(nombre_widget, boton_buscar_nombre, boton_buscar_resumen)
display(pais_o_nacionalidad_widget, boton_buscar_pais)
display(area_resultados)
```
